```python
from dotenv import load_dotenv
load_dotenv()
```

```python
PERPLEXITY_API_KEY= os.environ.get('PERPLEXITY_API_KEY')
```

```python
from openai import OpenAI
```

```python
messages = [
    {
        "role": "system",
        "content": (
            "You are a famous blogger who writes detail blogs and product reviews."
            "Before writing a blog you research about the product, it's product specifications and what other people review about it."
            "You consider everything and then write a product review about the product."
        ),
    },
    {
        "role": "user",
        "content": (
            "sony xm4"
        ),
    },
]
```

```python
client = OpenAI(api_key=PERPLEXITY_API_KEY, base_url="https://api.perplexity.ai")



# chat completion with streaming
response_stream = client.chat.completions.create(
    model="llama-3-sonar-large-32k-online",
    messages=messages,
    stream=True,
)

for response in response_stream:
    # print(response)
    print(response.choices[0].message['content'])
```

```python
# chat completion without streaming
response = client.chat.completions.create(
    model="llama-3-sonar-large-32k-online",
    messages=messages,
)
print(response)
```

```python
def create_streaming_blog(keyword: str):
    messages = [
            {
                "role": "system",
                "content": (
                    "You are a famous blogger who writes detail blogs and product reviews."
                    "Before writing a blog you research about the product, it's product specifications and what other people review about it."
                    "You consider everything and then write a product review about the product."
                ),
            },
            {
                "role": "user",
                "content": (
                    keyword
                ),
            },
        ]
    response_stream = client.chat.completions.create(
                model="llama-3-sonar-large-32k-online",
                messages=messages,
                stream=True,
            )
    return response_stream
    
```
